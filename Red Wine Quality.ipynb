## Adding column 'Reviews'
- It will make the visibility in good and bad by quality

```python
reviews = []
for i in df['Quality']:
    if i >= 3 and i <= 5:
        reviews.append('Bad')
    elif i >= 6 and i <= 8:
        reviews.append('Good')
df['Reviews'] = reviews
```

- Applying condition to make the difference visible
- If quality is greater than equal to 3 and less than equal to 5, it will consider it "Bad"
- If quality is greater than equal to 6 and less than equal to 8, it will consider it "Good" 

```python
df.head(10)
```

```python
df ['Reviews'].value_counts()
```

```python
sns.barplot(x = 'Reviews', y = 'Quality', data = df)

plt.title('Wine Quality by Reviews', fontsize = 14)
plt.xlabel('Quality', fontsize = 12)
plt.ylabel('Reviews', fontsize = 12)
plt.show()
```

- The differnce is completely visible
- The majority reviews are Good!

```python
avg_wine_quality = df.groupby('Quality')['Fixed_Acidity'].mean()

highest_fixed_acidity = avg_wine_quality.max()
lowest_fixed_acidity = avg_wine_quality.min()

print(f"Highest Fixed Acidity : {highest_fixed_acidity}")
print(f"Lowest Fixed Acidity : {lowest_fixed_acidity}")
```
